```python
from PIL import Image
import matplotlib.pyplot as plt
import os

normal_folder = "../data/chest_xray/train/NORMAL"

image_name = os.listdir(normal_folder)[0]

image_path = os.path.join(normal_folder, image_name)

img = Image.open(image_path)

plt.imshow(img, cmap="gray")
plt.axis("off")
plt.title("Normal Chest X-Ray")
plt.show()
```

```python
pneumonia_folder = "../data/chest_xray/train/PNEUMONIA"

image_name = os.listdir(pneumonia_folder)[0]

image_path = os.path.join(pneumonia_folder, image_name)

img = Image.open(image_path)

plt.imshow(img, cmap="gray")
plt.axis("off")
plt.title("Pneumonia Chest X-Ray")
plt.show()
```

```python
import os

normal_count = len(os.listdir("../data/chest_xray/train/NORMAL"))
pneumonia_count = len(os.listdir("../data/chest_xray/train/PNEUMONIA"))

print("Normal Images:", normal_count)
print("Pneumonia Images:", pneumonia_count)
```
